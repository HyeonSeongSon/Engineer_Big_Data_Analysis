```python
# 28 날짜와 시간, 문자열
# 1.연도와 월을 기준으로 주문 수를 집계하시오
# 2. 가장 많은 주문이 있었던 연도와 월을 예시와 같은 형식으로 숫자로만 구하시오

df = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/delivery_time.csv')

# 1
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['연월'] = df['주문시간'].dt.to_period('M')

# 2
# year_month = df['연월'].value_counts().idxmax()
year_month = df.groupby('연월').size().idxmax()

year_month = str(year_month)
year_month = year_month.replace("-",'')
year_month
```

```python
# 29 함수, 월별 집계
# 1. 배달료 계산 기준표에 따라 각 주문에 대한 배달료를 계산하시오
# 2. 월별로 배달료의 총합을 집계하시오
# 3. 배달료가 가장 맣이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 구하시오

df = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/delivery_time.csv')

# 1
def delivery_fee(distance):
  if distance < 5:
    return 2000
  elif distance < 10:
    return 4000
  elif distance < 15:
    return 6000
  elif distance < 20:
    return 8000

df['배달료'] = df['거리'].apply(delivery_fee)

# 2
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['월'] = df['주문시간'].dt.month
delivery_free_month_sum = df.groupby('월')['배달료'].sum()

# 3
delivery_free_month_sum.idxmax()
print(int(delivery_free_month_sum.max()))
```

```python
# 30 주말, 평일 구분
# 1. 주말 주문 건수와 평일 주문 건수를 구하시오
# 2. 주말 주문 건수와 평일 주문 건수의 차이를 절대값으로 구하고 정수형으로 구하시오

df = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/delivery_time.csv')

# 1
df['주문시간'] = pd.to_datetime(df['주문시간'])
df['dayofweek'] = df['주문시간'].dt.dayofweek
weekend = len(df[df['dayofweek'] < 5])
weekday = len(df[~(df['dayofweek'] < 5)])

# 2
print(abs(weekend - weekday))
```

```python
# 32 합계(열 방향), 상위 값 선택
# 1. 수학, 영어, 국어 점수합을 구하시오
# 2. 합이 가장 큰 상위 10명을 찾으시오
# 3. 찾은 10명의 수학 평균 점수를 구하시오 (반올림 후 정수 출력)

df = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/school_data.csv')

# 1
df['국영수'] = df[['수학','영어','국어']].sum(axis=1)

# 2
# top10 = df.sort_values('국영수', ascending=False).iloc[:10]
top10 = df.nlargest(10, '국영수')

# 3
print(round(top10['수학'].mean()))
```

```python
# 33 데이터프레임 재구조화
# 1. 과목에 상관없이 점수가 가장 작은 점수 25개를 찾으시오
# 2. 찾은 점수 25개의 합을 정수로 구하시오

df = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/school_data.csv')

# 1
melted_df = df.melt(id_vars=['이름'], value_vars=['수학','영어','국어'])
low25  = melted_df.nsmallest(25, 'value')

# 2
print(int(low25['value'].sum()))
```

```python
# 34 데이터 합치기(concat)
# 1. 두 파일 학생 순서는 같다
# 2. 학생별로 수학, 영어 국어, 과학 점수의 평균 구하시오
# 3. 평균 점수가 60점 이상인 인원 수를 계산하시오
df1 = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/school_data.csv')
df2 = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/school_data_science.csv')

# 1
df = pd.concat([df1, df2], axis=1)

# 2
df['평균'] = df[['수학','영어','국어','과학']].mean(axis=1)

# 3
print(sum(df['평균'] >= 60))
```

```python
# 35 데이터 합치기(merge)
# 1. 두 파일을 이름을 기준으로 합치시오
# 2. 영어교사가 장선생이면서 사회교사가 오선생인 학생들을 필터링하시오
# 3. 필터링된 학생들의 수학 점수를 모두 더한 후 정수로 구하시오
df1 = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/school_data.csv')
df2 = pd.read_csv('/content/drive/MyDrive/빅데이터분석기사파일/part1/ch3/school_data_social.csv')

# 1
df = pd.merge(df1, df2, on='이름')

# 2
df = df[(df['영어교사']=='장선생')&(df['사회교사']=='오선생')]

# 3
print(int(df['수학'].sum()))
```
